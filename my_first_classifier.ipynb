```python
import nltk 

from nltk.tokenize import word_tokenize 

from nltk.util import ngrams, skipgrams 

nltk.download('punkt') 

#Функция нахождения N_gram 

def get_ngrams(text, n): 

    text = text.lower().replace(',', '').replace('.', '').replace('(', '').replace(')', '') 

    words = word_tokenize(text, language="russian") 

    n_grams = ngrams(words, n) 

    return [ ' '.join(grams) for grams in n_grams] 

#Функция нахождения Skip-gram 

def get_skipgrams(text, n,k): 

    text = text.lower().replace(',', '').replace('.', '').replace('(', '').replace(')', '') 

    words = word_tokenize(text, language="russian") 

    skip_grams = skipgrams(words, n,k) 

    return [ ' '.join(grams) for grams in skip_grams] 

#Вывести список биграмм 
```

## Делаем из правильных данных n-gram и skip-gram 

```python
# n-gram
n=2
n_gramms=[] 
for card in test['description']:
    n_gramms+=get_ngrams(card, n)
```

```python
nfdist = nltk.FreqDist(n_gramms)
print(nfdist)
```

```python
# s-gram
n=3
k=2
s_gramms=[] 
for card in test['description']:
    s_gramms+=get_skipgrams(card, n, k)
```

```python
sfdist = nltk.FreqDist(s_gramms)
print(sfdist)
```
